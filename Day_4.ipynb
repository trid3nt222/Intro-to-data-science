WHILE loop

```python
#Simple program that prints "Hello World" 5 times
counter=5 #hold the no. of iterations
while counter>0:
  print("Hello World")
  counter=counter-1
```

```python
counter=0
while counter<5:
  print("Hello World")
  counter=counter+1
```

```python
counter=1
while counter<=5:
  print("Hello World")
  counter=counter+1
```

```python
x="***"
counter=0
while counter<3:
  print(x)
  counter=counter+1
  x=x+x

```

```python
counter=1
while counter<=3:
  print("***"*counter) #multiplying str and an int-the str is repeated
  counter=counter+1
```

```python
password_incorrect=True
password="secret"
while password_incorrect:
  user_pass=input("Enter password: ")
  if user_pass==password:
    print("Welcome back. You are logged in")
    password_incorrect=False #control
  else:print("Password incorrect! Try again")
```

```python
age_not_found=True
while age_not_found:
  age=input("What is your age: ")
  if age=="18":
    print("Age found")
    age_not_found=False
```

```python
letters="xx"
space=" "
counter=7
while counter>0:
  print((space*(7-counter))+letters+(letters*counter))
  counter=counter-1
```
