# FOR loop

```python
#rage(5)-starts from 0,stops before 5
for number in range(5):
  print(number)
```

```python
#range(1,6)-starts automatically from 1,stops before 6
for number in range(1,6):
  print(number)
```

```python
#range(2,11,2)-starts from 2,stops before 11,increments(skips) by 2
for number in range(2,11,2):
  print(number)
```

```python
#slicing backwards
for number in range(10,0,-1):
  print(number)
```

```python
for number in range(1,11):
  print(f"5x{number}={5*number}")
```

```python
for i in range(1,11):
  for j in range(1,11):
    print(f"{i}x{j}={i*j}")
  print()#blank line after each table
```

```python
print("\t", end="")

#Print column headings
for i in range(1,11):
  print(i,end="\t")

print()

#print rows
for i in range(1,11):
  print(i, end="\t") #row heading
  for j in range(1,11):
    print(i*j, end="\t")
  print()
```

```python
print("\t", end="")

#Print column headings
for i in range(3,11):
  print(i,end="\t")

print()

#print rows
for i in range(3,11):
  print(i, end="\t") #row heading
  for j in range(3,11):
    print(i*j, end="\t")
  print()
```

```python
for number in range(10):
  if number==5:
    break
  print(number)
```

```python
for number in range(10):
  if number==5:
    continue
  print(number)
```
