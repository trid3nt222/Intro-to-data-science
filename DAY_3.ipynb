# Boolean Operators
and-all conditions must be true

or-at least one condition must be true

````python
a=1
b=2
c=3
print(a==b and b==c)
print(a<b or b>c)
print(a<b and b>c)
````

Write a program that takes a value from the user, the program should print

*divisible by two if the value is divisible by two

*divisible by three if the value is divisible by three

*divisible by both two and three is the value is divisible by both two and three

otherwise print not divisible by both 2 and 3

````python
value=int(input("Enter value: "))
if value%3==0 and value%2==0:
  print("divisible by both 2 and 3")
elif value%2==0:
  print("divisible by 2")
elif value%3==0:
  print("divisible by 3")

else:
  print("not divisible by both 2 and 3")
````
