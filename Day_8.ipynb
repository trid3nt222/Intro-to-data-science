````python
def authenticator(password):
  defined_pass="secret"
  if password==defined_pass:
    return True
  else:
    return False
````

````python
passw=input("Enter password: ")
if authenticator(passw):
  print("You are logged in")
else:
  print("Wrong password")
````

````python
def sum_diff(x,y):
  s=x+y
  d=x-y
  return(s,d)
````

````python
sum_of_both, diff_of_both=sum_diff(100,50)
print(sum_of_both)
print(diff_of_both)
````
