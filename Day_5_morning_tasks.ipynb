```python
#72-palindrome
word=input("Word: ")
palindrome=True
left=0
right=len(word)-1
while left<right:
  if word[left]!=word[right]:
    palindrome=False
    break
  left=left+1
  right=right-1
  if palindrome:
    print("palindrome")
  else:
    print("is not a palindrome")
```

```python
#GCD
n=float(input("Enter n: "))
m=float(input("Enter m: "))
d=min(m,n)
while d>0:
     if (n%d)!=0 or (m%d)!=0:
      d=d-1
     else:
      print("GCD:",d)
      d=0
```

```python
#70-Caesar Cipher
alphabet = "abcdefghijklmnopqrstuvwxyz"
choice = input("Enter E for encoding or D for decoding: ").upper()
letter = input("Enter a letter: ").lower()
shift = 3
message = ""
while letter != "stop":
    if letter in alphabet:
        index = alphabet.index(letter)
        if choice == "E":
            new_index = (index + shift) % 26
        elif choice == "D":
            new_index = (index - shift) % 26
        else:
            print("Invalid choice!")
            break
        message += alphabet[new_index]
    elif letter == " ":
        message += " "
    else:
        print("Invalid input")
    letter = input("Enter the next letter (or type 'stop' to finish): ").lower()
if choice == "E":
    print("Encrypted message:", message)
elif choice == "D":
    print("Decoded message:", message)
```
