#Dictionaries
* Collection of key:value pairs
* Mutable
* Mixed datatypes

*syntax*

variable={key:value}

```python
student_grades={
  "math":80,
  "english":95,
  "cre":92
}
print(type(student_grades))
```

```python
math_results=student_grades["math"]
print(math_results)
```

```python
#creating/adding a value
student_grades["physics"]=78
print(student_grades)
#creates/adds physics
```

```python
#changing a value
student_grades["math"]=82
print(student_grades)
#changes math
```

```python
#deleting a value
del student_grades["cre"]
```

```python
print(student_grades)
```

```python
#updating/adding values- .update({dict})
student_grades.update({"math":70,"cre":90})
print(student_grades)
```

```python
print(student_grades["math"])
```

```python
print(student_grades.get("math",0))
```

```python
student_grades.pop("math")#deleting an item-.pop(key)
```

```python
print(student_grades.get("math",0))
```

```python
#calling the keys- .keys()
print(student_grades.keys())
```

```python
#calling the values- .values()
print(student_grades.values())
```

```python
#calling both the keys&values- .items()
#returns as tuples
print(student_grades.items())
```

```python
for key in student_grades.keys():
  print(key.upper())
```

```python
for val in student_grades.values():
  print(val)
```

```python
for item in student_grades.items():
  print(item)
```

```python
for key,value in student_grades.items():
  print(key,"contains",value)
```
