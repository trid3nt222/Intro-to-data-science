```python
data="""name,gender,grade
John,m,76
Jane,f,82
Joseph,m,84
Susan,f,72
"""
rows=data.split("\n")
cols=rows[0].split(",")
print(cols)
```

```python
#.append()-adds item at the end
cols.append("age")
print(cols)
```

```python
#.pop()-removes item at the end
cols.pop(1)
print(cols)
```

```python
#.insert()-allows insertion of a given index
cols.insert(1,"weight")
```

```python
#.remove()-allows removal of am item in lost
if "weight" in cols:
  cols.remove("weight")
```

```python
#.extend()-joins two or more lists
cols.extend(["weight","age"])
```

.clear()-removes all items

.index()-get index of an item

.count()-find no. of occurences of an item

.sort()-arranges from smallest to largest

.reverse()-reverses your list

```python
#from smallest to largest
marks=[20,34,18,45,4,7]
marks.sort()
print(marks)
```

```python
#from largest to smallest
marks.sort(reverse=True)
print(marks)
```
