# WHILE LOOPS

```python
#Prints numbers from 1-10
num=1
while num<=10:
  print(num)
  num+=1 #control
```

```python
#Write a program that counts down from 10 to 1
num=10
while num>=1:
  print(num)
  num-=1 #control
```

```python
#Write a program that asks the user for a password until they enter "python"
incorrect_password=True
correct_password="python"
while incorrect_password:
  password=input("Enter password: ")
  if password==correct_password:
    print("Access granted!")
    incorrect_password=False #control
  else:
    print("Incorrect password. Try again")


```

```python
#Write a program that keeps asking the user for a positive number.
negative=True
while negative:
  num=int(input("Enter a no: "))
  if num>0:
    print("positive")
    negative=False #control
  else:
    print("negative. Try again")

```

Write a Python program that has an account balance of Ksh 10,000.

The program should:

Ask the user how much money they want to withdraw.

If the amount is more than the balance, display:

Insufficient funds.

and ask again.

If the amount is less than or equal to the balance, subtract it from the balance.
Display the new balance.
Continue asking the user if they want to make another withdrawal (yes or no).
The program should stop when the user enters no.

```python
acc_bal=10000
while True:
  amt=float(input("How much do you want to withdraw: "))
  if amt>acc_bal:
    print("Insufficient balance")
    continue
  else:
    acc_bal=acc_bal-amt
    print("Your balance:",acc_bal)
    question=input("Do you want to make another withdrawal?")
  if question!="yes":
    break
```
