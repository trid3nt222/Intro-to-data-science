```python
#Entering student details
students = []
more = "yes"
total = 0
while more == "yes":
    name = input("Enter student name: ")
    age = input("Enter student age: ")
    gender = input("Enter student gender: ")

    student = {
        "name": name,
        "age": age,
        "gender": gender,
    }

    students.append(student)

    more = input("Add another student? (yes/no): ")

#recording marks
marks = []
for i in range(len(students)):
  print(f"Student {i + 1} marks:")
  mark = input("Enter marks: ")
  students[i]["mark"] = mark

print("=============STUDENT DETAILS================")
index=0
while index < len(students):
  print("Student", index + 1)
  print("Name:", students[index]["name"])
  print("Age:", students[index]["age"])
  print("Gender:", students[index]["gender"])
  print("Marks:", students[index]["mark"])
  print("\n")
  index = index + 1

#viewing failing scores
print("FAILING SCORES")
fail=False

for i in students:
  if int(i["mark"]) < 50:
    print(i["name"],"-",i["mark"])
    fail=True

if fail==False:
  print("No failing scores")
  print("\n")

print("Program exit")
```
